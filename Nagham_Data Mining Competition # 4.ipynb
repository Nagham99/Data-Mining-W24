### Answering the Questions:

ðŸŒˆ **Is a fully-connected model a good one for sequential data? How about for image data?**

**Sequential Data:** No, fully-connected models are not ideal for sequential data because they lack the ability to capture temporal dependencies. RNNs, especially variants like LSTM or GRU, are more suitable for sequential data due to their ability to retain memory.

**Image Data:** Fully-connected models can perform well for image data, especially when combined with convolutional layers. However, for tasks requiring understanding of spatial relationships, architectures like CNNs are more effective as they can capture patterns in image data.

ðŸŒˆ **What is gradient vanishing and gradient explosion, and how do GRU/LSTM mitigate this problem?**

**Gradient Vanishing:** It occurs when gradients become very small during backpropagation, leading to slow learning or convergence. 

**Gradient Explosion:** It occurs when gradients become excessively large, leading to unstable training.

**GRU/LSTM Mitigation:** GRU and LSTM mitigate these problems by using gating mechanisms that control the flow of information. By regulating the flow of gradients through the cell state, they can prevent vanishing gradients and stabilize the training process.

ðŸŒˆ **What is multi-objective/multi-task learning? What is multi-modality learning? How do you use them in this assignment?**

**Multi-Objective/Multi-Task Learning:** Involves optimizing multiple objectives or tasks simultaneously during training. This can improve generalization and learning efficiency by sharing information between related tasks.

**Multi-Modality Learning:** Deals with learning from multiple modalities of data (text, images, etc.) and combining them for better understanding or prediction.

**Usage in Assignment:** In this assignment, multi-objective/multi-task learning can be used to predict the pricing range (beginner, plus, premium) while simultaneously optimizing other objectives. Multi-modality learning can be applied by incorporating various features (listing characteristics) such as text descriptions, image thumbnails, etc., to enhance prediction accuracy.

ðŸŒˆ **What is the difference among xgboost, lightgbm, and catboost?**

- **XGBoost (eXtreme Gradient Boosting):** The original gradient boosting algorithm, known for its effectiveness. It uses a level-wise tree growth strategy.

- **LightGBM:** Optimized for speed and efficiency, it uses a leaf-wise tree growth strategy. It's faster and more memory-efficient than XGBoost, especially for large datasets.

- **CatBoost:** Specifically designed to handle categorical features efficiently, with robust handling of missing data. It internally deals with categorical variables and has built-in support for GPU training.

Each algorithm has its own strengths and is suitable for different scenarios, but all are gradient boosting algorithms used for supervised learning tasks.

## Text Data preprocessing 

```python
from googletrans import Translator
import pandas as pd
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Initialize tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_data['summary'])
max_seq_length = max(len(seq) for seq in tokenizer.texts_to_sequences(train_data['summary']))

# Initialize translator
translator = Translator()

# Function to translate non-English summaries to English
def translate_to_english(text):
    try:
        translation = translator.translate(text, dest='en')
        return translation.text
    except:
        return text

# Apply translate_to_english function to summary column after tokenizing
train_data['summary'] = tokenizer.texts_to_sequences(train_data['summary'])
train_data['summary'] = train_data['summary'].apply(translate_to_english)
train_data['summary'] = pad_sequences(train_data['summary'], maxlen=max_seq_length, padding='post')
```

## Trail 1: LSTM with CNN for Text Input

#### Model Architecture:
- **Input Layer:** An embedding layer converts text data into dense word vectors. The embedding layer is followed by an LSTM layer with 64 units.
- **Output Layer:** A dense layer with softmax activation predicts the price category (assuming 3 output classes).

#### Thoughts and Observations for Trial:

**Thoughts:**

- Training loss decreases over epochs, but the validation loss fluctuates, indicating that the model may not be generalizing well.

Plan for next trail: Text Input with Bidirectional LSTM

```python
# Tokenize the text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_data['summary'])
X = tokenizer.texts_to_sequences(train_data['summary'])

# Pad sequences to ensure uniform length
X = pad_sequences(X)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X, train_data['price'], test_size=0.2, random_state=42)

# Define the model
model = Sequential()
model.add(Embedding(len(tokenizer.word_index) + 1, 128))
model.add(LSTM(64))
model.add(Dense(3, activation='softmax'))

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
# Train the model
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))
```

```python
# Model Evaluation
val_predictions = model.predict(X_val)
val_predictions_classes = np.argmax(val_predictions, axis=1)
val_accuracy = accuracy_score(y_val, val_predictions_classes)
print("Validation Accuracy:", val_accuracy)
```

```python
# Make predictions on the test dataset
test_sequences = tokenizer.texts_to_sequences(test_data['summary'])
X_test = pad_sequences(test_sequences, maxlen=X.shape[1])

test_predictions = model.predict(X_test)
test_predictions_classes = np.argmax(test_predictions, axis=1)

# Create the submission dataframe
submission = pd.DataFrame({'id': test_data['id'], 'price': test_predictions_classes})
# Save the submission to a CSV file
submission.to_csv('Text Classification with LSTM.csv', index=False)
```

## Trail 3: Bidirectional GRU for Text Classification

### Model Architecture:
- The model consists of an Embedding layer, followed by a Bidirectional GRU layer, and a Dense layer with softmax activation for classification.
- The Embedding layer converts integer-encoded words into dense vectors of fixed size.
- Bidirectional GRU layer processes the input sequences in both forward and backward directions to capture contextual information effectively.
- The Dense layer with softmax activation outputs the predicted class probabilities.

### Thoughts and Observations

#### Thoughts:
- Bidirectional GRU is expected to capture both past and future information, which can be beneficial for understanding the context of the text data.
- Since the model architecture includes bidirectional processing, it may require more computational resources compared to unidirectional models.
- Bidirectional models have the potential to perform better than unidirectional ones, especially in tasks where context plays a crucial role.

#### Observations :
- The model shows moderate performance with an accuracy of around 62% on the validation set.
- The training and validation loss seem to plateau after a few epochs, suggesting that the model may not benefit significantly from further training.
- The accuracy on the validation set is slightly higher than the baseline, but it does not show significant improvement.

```python
from tensorflow.keras.layers import Embedding, Bidirectional, GRU, Dense

# Define the model
modelGRU = Sequential([
    Embedding(len(tokenizer.word_index) + 1, 128),
    Bidirectional(GRU(64)),
    Dense(3, activation='softmax')
])

# Compile the model
modelGRU.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = modelGRU.fit(X_train, y_train, epochs=10, batch_size=32, validation_data=(X_val, y_val))

# Model Evaluation
val_loss, val_accuracy = modelGRU.evaluate(X_val, y_val)
print("Validation Accuracy:", val_accuracy)

# Make predictions on the test dataset
test_sequences = tokenizer.texts_to_sequences(test_data['summary'])
X_test = pad_sequences(test_sequences, maxlen=X.shape[1])

test_predictions = modelGRU.predict(X_test)
test_predictions_classes = np.argmax(test_predictions, axis=1)
```

```python
# Create the submission dataframe
submission = pd.DataFrame({'id': test_data['id'], 'price': test_predictions_classes})

# Save the submission to a CSV file
submission.to_csv('text_GRU.csv', index=False)
```

```python

from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Embedding, LSTM, concatenate, Input

# Drop rows with missing data
train_data.dropna(subset=['summary', 'image'], inplace=True)

# Tokenize text data
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_data['summary'])
text_sequences = tokenizer.texts_to_sequences(train_data['summary'])

# Pad text sequences to the same length
max_seq_length = max(len(seq) for seq in text_sequences)
text_data = pad_sequences(text_sequences, maxlen=max_seq_length)

# Split the data into training and validation sets
X_img_train, X_img_val, X_text_train, X_text_val, y_train, y_val = train_test_split(x_train_image, text_data, train_data['price'], test_size=0.2, random_state=42)

# Define the image input
img_input = Input(shape=(64, 64, 2), name='img_input')
conv1 = Conv2D(32, (3, 3), activation='relu')(img_input)
pool1 = MaxPooling2D((2, 2))(conv1)
conv2 = Conv2D(64, (3, 3), activation='relu')(pool1)
pool2 = MaxPooling2D((2, 2))(conv2)
flatten_img = Flatten()(pool2)

# Define the text input
text_input = Input(shape=(max_seq_length,), name='text_input')
embedding = Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=max_seq_length)(text_input)
lstm = LSTM(64)(embedding)

# Concatenate image and text features
concatenated = concatenate([flatten_img, lstm])

# Dense layers for combined features
dense1 = Dense(64, activation='relu')(concatenated)
output = Dense(3, activation='softmax')(dense1)  # Assuming 3 output classes for price categories

# Create the model
model = Model(inputs=[img_input, text_input], outputs=output)

# Compile the model
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit([X_img_train, X_text_train], y_train, epochs=10, batch_size=32, validation_data=([X_img_val, X_text_val], y_val))


```

## Trail 7: Multi-objective Learning for Price and Type Prediction
### Model Architecture

The multi-objective learning model architecture consists of two main branches for processing image and text data, respectively. Here is a detailed breakdown of the model architecture:

#### Image Input Branch:
- Input: Images with dimensions 64x64x2 (grayscale).
- Convolutional Layers: Two Conv2D layers with 32 and 64 filters, respectively, each followed by ReLU activation functions.
- Pooling Layers: Two MaxPooling2D layers to reduce spatial dimensions by half.
- Flattening: The output is flattened to a one-dimensional vector.
  
#### Text Input Branch:
- Input: Text summaries tokenized and padded to a maximum sequence length.
- Embedding Layer: Converts each token into dense vectors of fixed size (100 dimensions).
- LSTM Layer: Captures sequential dependencies in the text data with 64 units.

#### Combined Features:
- Concatenation: The features extracted from both branches are concatenated.
- Dropout: Dropout regularization (dropout rate of 0.5) is applied to prevent overfitting.

#### Output Layers:
- Price Output: Dense layer with softmax activation for predicting the price category (3 classes).
- Type Output: Dense layer with softmax activation for predicting the type category.

#### Overall Model:
- The model is compiled with the Adam optimizer.
- Loss Functions: Separate sparse categorical cross-entropy loss functions are used for price and type predictions.
- Metrics: Accuracy is used as the evaluation metric for both price and type predictions.



**Thoughts:**

- Multi-objective learning aims to predict both price and type simultaneously, which can be advantageous in capturing more information from the dataset.
- The model architecture includes separate output layers for price and type prediction, each with its own loss function.
- It's essential to monitor the performance of both objectives individually to ensure balanced learning.
- Dropout regularization has been added to prevent overfitting and improve generalization.
- The training loss and accuracy for both price and type predictions need to be analyzed to understand the model's learning process.

**Observations:**

- Compared to the previous trail with only image input, the multi-objective model shows better performance in terms of validation accuracy for both price and type prediction.
- The model achieved decent accuracy for predicting the type but struggled with price prediction, similar to the previous trail.
- The training loss for both objectives decreased over epochs, indicating learning, but the validation loss did not decrease consistently, which is also similar to the previous trail.
- There is a noticeable gap between training and validation accuracy, indicating some degree of overfitting, similar to the previous trail.
- The accuracy for predicting type is higher compared to price, suggesting that the model may need further tuning or additional data to improve price prediction performance.

```python
train_data.dropna(subset=['summary', 'image', 'type', 'price'], inplace=True)

# Tokenize text data
from tensorflow.keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer()
tokenizer.fit_on_texts(train_data['summary'])
text_sequences = tokenizer.texts_to_sequences(train_data['summary'])
max_seq_length = max(len(seq) for seq in text_sequences)
text_data = pad_sequences(text_sequences, maxlen=max_seq_length)

# Encode the 'type' column
type_encoder = LabelEncoder()
type_data = type_encoder.fit_transform(train_data['type'])

# Split the data into training and validation sets
X_img_train, X_img_val, X_text_train, X_text_val, y_price_train, y_price_val, y_type_train, y_type_val = train_test_split(
    x_train_image, text_data, train_data['price'], type_data, test_size=0.2, random_state=42
)

# Define the image input
img_input = Input(shape=(64, 64, 2), name='img_input')
conv1 = Conv2D(32, (3, 3), activation='relu')(img_input)
pool1 = MaxPooling2D((2, 2))(conv1)
conv2 = Conv2D(64, (3, 3), activation='relu')(pool1)
pool2 = MaxPooling2D((2, 2))(conv2)
flatten_img = Flatten()(pool2)

# Define the text input
text_input = Input(shape=(max_seq_length,), name='text_input')
embedding = Embedding(input_dim=len(tokenizer.word_index)+1, output_dim=100, input_length=max_seq_length)(text_input)
lstm = LSTM(64)(embedding)

# Concatenate image and text features
concatenated = concatenate([flatten_img, lstm])
# Adding Dropout
concatenated = Dropout(0.5)(concatenated)

# Dense layers for combined features (Price)
dense_price = Dense(64, activation='relu')(concatenated)
output_price = Dense(3, activation='softmax', name='price_output')(dense_price)  # Output layer for price prediction

# Dense layers for combined features (Type)
dense_type = Dense(64, activation='relu')(concatenated)
output_type = Dense(len(type_encoder.classes_), activation='softmax', name='type_output')(dense_type)  # Output layer for type prediction

# Create the model
model = Model(inputs=[img_input, text_input], outputs=[output_price, output_type])

# Compile the model
model.compile(optimizer='adam',
              loss={'price_output': 'sparse_categorical_crossentropy', 'type_output': 'sparse_categorical_crossentropy'},
              metrics={'price_output': 'accuracy', 'type_output': 'accuracy'})

# Train the model
model.fit({'img_input': X_img_train, 'text_input': X_text_train},
          {'price_output': y_price_train, 'type_output': y_type_train},
          epochs=10, batch_size=32,
          validation_data=({'img_input': X_img_val, 'text_input': X_text_val},
                           {'price_output': y_price_val, 'type_output': y_type_val}))
```

```python
# Model Evaluation
val_results = model.evaluate({'img_input': X_img_val, 'text_input': X_text_val},
                              {'price_output': y_price_val, 'type_output': y_type_val})

print("Validation Results:", val_results)

# Tokenize and pad text data for test set
test_text_sequences = tokenizer.texts_to_sequences(test_data['summary'])
test_text_data = pad_sequences(test_text_sequences, maxlen=max_seq_length)

price_predictions, type_predictions = model.predict({'img_input': x_test_image, 'text_input': test_text_data})

# Create the submission dataframe
subm = pd.DataFrame({
    'id': test_data['id'],
    'price': np.argmax(price_predictions, axis=1),  # Convert softmax probabilities to class labels
    'type': type_encoder.inverse_transform(np.argmax(type_predictions, axis=1))
})
```
